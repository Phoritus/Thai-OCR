## Qwen3.5:9B

```python
from ollama import chat

prompt = """
You are an expert data extraction assistant specialized in analyzing Thai receipts and bills. 
Extract the key information from the provided image and format your entire response strictly as valid JSON. 
Do not include any explanations, greetings, or markdown formatting.
"""

response = chat(
    # Make sure to use a vision-capable model!
    model='qwen3.5:9b', 
    messages=[{
        'role': 'user',
        'content': prompt,
        'images': ['p1.jpg']  # <-- Add your image path(s) here in a list
    }],
)

print(response.message.content)
```
